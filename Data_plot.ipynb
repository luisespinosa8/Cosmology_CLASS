```python
import numpy as np
import os
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
bkg = np.genfromtxt('bkg_data')
H = bkg[3,:]
z = bkg[0,:]
rho_g = bkg[8,:]
rho_b = bkg[9,:]
rho_cdm = bkg[10,:]
rho_lambda = bkg[11,:]
rho_tot = bkg[13,:]
```

```python
mPk = np.genfromtxt('/Users/luisenriqueespinosacastro/class/output/output_mPk')
k = mPk[0,:]
Pk = mPk[1,:]
```

```python
cmb = np.genfromtxt('cmb_data')
cl_tt = cmb[1,:]
cl_ee = cmb[2,:]
l = cmb[0,:]
```

```python
H_norm = H / (1+z)
a = 1 / (z+1)
```

```python
plt.figure()
plt.plot(z[0:100],H_norm[0:100])
plt.title('Hubble parameter evolution',fontsize=20)
#plt.yscale('log')
#plt.xscale('log')
plt.grid()
plt.xlabel('z',fontsize=20)
plt.ylabel(r'H(z)/1+z',fontsize=20)
plt.savefig('H_plot.png')
```

```python
plt.figure()
plt.plot(a,H)
plt.title('Hubble parameter evolution',fontsize=20)
plt.yscale('log')
plt.xscale('log')
plt.axvline(x=1/(3400+1),color='black',linestyle='--')
plt.axvline(x=1/(0.3+1),color='black',linestyle='--')
plt.grid()
plt.xlabel('a',fontsize=20)
plt.ylabel(r'$H(a)$',fontsize=20)
plt.savefig('H_plot2.png')
```

```python
plt.figure()
plt.plot(a, rho_g, label='Radiation')
#plt.plot(a, rho_b, label='Baryonic matter')
#plt.plot(a, rho_cdm, label='Dark matter')
plt.plot(a, rho_b+rho_cdm, label='Matter')
plt.plot(a, rho_lambda, label='Dark energy')
plt.plot(a, rho_tot-(rho_b+rho_cdm+rho_lambda), label='Total')
plt.axvline(x=1/(3400+1),color='black',linestyle='--')
plt.axvline(x=1/(0.3+1),color='black',linestyle='--')
plt.title('Density evolutions',fontsize=20)
plt.ylim(1e-15, 1e55)
plt.yscale('log')
plt.xscale('log')
plt.legend(bbox_to_anchor=(1, 0.5))
plt.grid()
plt.xlabel('a',fontsize=20)
plt.ylabel(r'$\rho(z)$',fontsize=20)
plt.savefig('rho_plot.png')
```

```python
plt.figure()
plt.plot(k,Pk)
plt.title('Power spectrum',fontsize=20)
plt.yscale('log')
plt.xscale('log')
plt.grid()
plt.xlabel('k',fontsize=20)
plt.ylabel('P(k)',fontsize=20)
plt.savefig('Pk_plot.png')
```

```python
plt.figure()
plt.plot(l,cl_tt)
plt.title('Cosmic Microwave Background',fontsize=20)
#plt.yscale('log')
#plt.xscale('log')
plt.grid()
plt.xlabel('l',fontsize=20)
plt.ylabel(r'$C_{TT}(k)$',fontsize=20)
plt.savefig('Pk_plot.png')
```

```python
plt.figure()
plt.plot(l,cl_ee)
plt.title('Cosmic Microwave Background',fontsize=20)
#plt.yscale('log')
#plt.xscale('log')
plt.grid()
plt.xlabel('l',fontsize=20)
plt.ylabel(r'$C_{EE}(l)$',fontsize=20)
plt.savefig('Pk_plot.png')
```
